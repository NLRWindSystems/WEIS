**Key Points:**

- **OpenFAST Integration**: Full aeroelastic simulation with platform dynamics

````python
# Setting up file paths
stageFolder = "stage-4-dlcs"

use_precomputed = True

if use_precomputed:
    sum_stat_file = os.path.join(this_dir,"..","1_drivers", stageFolder,"outputs_preCompute/dlcs/rank_0/iteration_0/summary_stats.p")
else:
    sum_stat_file = os.path.join(this_dir,"..","1_drivers", stageFolder,"outputs/dlcs/rank_0/iteration_0/summary_stats.p")


with open(sum_stat_file, 'rb') as f:
    data = pickle.load(f)

custom_y_params = [
            ('GenSpeed', 'mean'),
            ('RootMyb1', 'max'),
            ('TwrBsAxMxyt', 'median'),
            ('PtfmPitch', 'max'),
]

plot_weis_summary(data, custom_y_params)
````
